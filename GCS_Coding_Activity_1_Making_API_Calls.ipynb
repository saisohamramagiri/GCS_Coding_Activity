```python
import requests
```

```python
class Weather:
    def __init__(self, api_key):
        self.api_key = api_key
        self.favorites = list()

    # 1. Search for Weather Details of a City: Enter the name of a city and display its current weather details using the OpenWeather API.
    def search_weather_details(self, city):
      url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={self.api_key}&units=imperial"
      response = requests.get(url)
      # print(response)
      # print(reponse.status_code)
      if response.status_code == 200:     # 200 means OK or the Request works fine whereas, 404 means Request Not Found
        x = response.json()
        y = x['main']
        current_temp = y['temp']
        weather_data = x['weather'][0]
        print(f"Weather details for the requested city \"{city}\" are as follows:")
        print("Temperature:", current_temp, "°F")
        print(f"Forecast: {weather_data['description']}")
      else:
        print(f"Error: Unable to service the request to fetch weather details for {city}.")

    # 2. Add a City to Favorites: Allow users to add a city to their list of favorites, with a maximum of three cities.
    def add_city_to_favorites(self, city):
      if len(self.favorites) <= 2:
        self.favorites.append(city)
        print(f"{city} has been added to your favorites.")
      else:
        print("Error: Favorite cities cannot exceed the Maximum Limit of 3.")

    # 3. List Favorite Cities: Display the list of favorite cities along with their current weather details.
    def list_favorites(self):
      print("Your Favorite Cities and their Weather Forecast are as follows:")
      for city_name in self.favorites:
        print(city_name, ":")
        self.search_weather_details(city_name)
        print()

    # 4. Update Favorite Cities: Enable users to remove a city from their favorites and add a new one, ensuring the limit of three cities at a time.
    def update_favorites(self, remove_city, add_city):
      if remove_city in self.favorites:
        self.favorites.remove(remove_city)
        self.add_city_to_favorites(add_city)
      else:
        print(f"Error: {remove_city} not found in your favorites list. Please re-check your list, and try again!")
```

```python
def user_interface():
  while True:
        print("\nUser Menu:")
        print("1. Search for Weather Details of a City")
        print("2. Add a City to Favorites")
        print("3. List all Favorite Cities")
        print("4. Update Favorite Cities")
        print("5. Exit")
        choice = input("Enter a number to indicate your choice (1-5): ")

        if choice == '1':
            city = input("Enter the name of the city to fetch its weather details: ")
            weather.search_weather_details(city)
        elif choice == '2':
            city = input("Enter the name of the city to add to favorites: ")
            weather.add_city_to_favorites(city)
        elif choice == '3':
            weather.list_favorites()
        elif choice == '4':
            remove_city = input("Enter the name of the city to remove from favorites: ")
            add_city = input("Enter the name of the city to add to favorites: ")
            weather.update_favorites(remove_city, add_city)
        elif choice == '5':
            print("Thank you for using the Weather App!")
            break
        else:
            print("Please enter a number between 1 and 5.")
```
